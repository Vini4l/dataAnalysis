```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
x = [0.205,0.196,0.182,0.189,0.197]
y = [202.60, 303.60, 402.60, 502.60, 595.93]
xerr = 0.009
yerr = 0

fig, ax = plt.subplots()


ax.errorbar(x, y,
            xerr=xerr,
            yerr=yerr, capsize=5,
            fmt='ok')


ax.set_xlabel('$\mu_{e}$')
ax.set_ylabel('$Massa(kg)$')
ax.set_title('')

plt.show()
plt.savefig('error.png')
```
